```python
class Square:
    def __init__(self):
        self.square = 0
        while True :  # 프로그램이 종료될 때 까지 반복되도록 While roop 생성
            self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴, 종료하려면 0을 누르세요 \n >>>'))

            if self.square == 1:
                print('직사각형 함수는 rect()입니다.\n')
                print(self.rect())  # 바로 rect 메소드가 실행되도록

            elif self.square == 2:
                print('평행사변형 함수는 par()입니다.\n')
                print(self.par())  # 바로 par 메소드가 실행되도록

            elif self.square == 3:
                print('사다리꼴 함수는 trape()입니다.\n')
                print(self.trape())  # 바로 trape 메소드가 실행되도록

            elif self.square == 0:
                print('프로그램을 종료합니다')
                break   # 0을 입력하였을 때만 무한루프 탈출하면서 종료하기

            else:
                print('1, 2, 3 중에서 다시 입력해주세요\n') # 0,1,2,3 외 다른숫자 입력 시 다시 입력하도록 복귀

    def rect(self): #직사각형 넓이 구하는 메소드
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = f"직사각형의 넓이는 {area} 입니다.\n"
        return result

    def par(self): #직사각형 넓이 구하는 메소드
        width, length = map(int, input("가로, 세로의 길이를 입력하세요. 예시: 가로,세로\n>>>").split(','))
        area = width * length
        result = f"평행사변형의 넓이는 {area} 입니다.\n"
        return result

    def trape(self): #직사각형 넓이 구하는 메소드
        width1, width2, length = map(int, input("가로1, 가로2, 세로의 길이를 입력하세요. 예시: 가로1, 가로2 ,세로\n>>>:").split(','))
        area = (width1+width2) * length * 0.5
        result = f"사다리꼴의 넓이는 {area} 입니다.\n"
        return result

a = Square()
```

```python
class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.temp = ['HOT', 'ICE']
        self.order_list = []
        self.order_price = []
        self.price_sum = 0

    def start(self):
        print("주문을 시작합니다\n")
        self.menu_print()

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(f"{i+1}.{self.menu[i]} : {self.price[i]}원")
        self.menu_select()  # 바로 주문메서드로 연결

    # 주문 메서드
    def menu_select(self):
        n = 0  # n값 초기화
        while n < 1 or 6 < n: # 메뉴를 선택할 때 까지 반복
            n = int(input("\n메뉴 번호를 입력하세요 >>> "))

            if 1 <= n <= 6:
                self.order_price.append(self.price[n-1])
                self.price_sum += self.price[n-1]
            else:
                print("\n잘못된 입력입니다. 다시 입력하세요\n")

        t = 0  # p값 초기화
        while t != 1 and t != 2: # 온도를 선택할 때 까지 반복
            t = int(input(f"\n원하는 음료의 온도를 선택하세요\n 1. HOT {self.menu[n-1]}, 2. ICE {self.menu[n-1]} >>> "))

            if t == 1 or t == 2:
                self.order_list.append(f"{self.temp[t-1]} {self.menu[n-1]}") # '온도 메뉴'항목이 주문리스트에 포함
            else :
                print("\n잘못된 입력입니다. 1 또는 2를 입력하세요.\n")

        print(f"\n고르신 메뉴는 {self.temp[t-1]} {self.menu[n-1]} : {self.price[n-1]}원 입니다.\n") # 메뉴 재확인하기

        #추가주문
        while n != 0:  # 지불을 선택하기 전까지 반복
            n = int(input("추가주문을 하시려면 메뉴 번호를 입력하세요. 지불을 원하시면 0을 눌러주세요 >>> "))
            if n > 0 and n < len(self.menu) + 1:
                self.order_price.append(self.price[n-1])  # 가격 리스트에 추가합니다.
                self.price_sum += self.price[n-1]  # 합계 금액

                t = 0
                while t != 1 and t != 2:
                    t = int(input(f"\n원하는 음료의 온도를 선택하세요\n 1. HOT {self.menu[n-1]}, 2. ICE {self.menu[n-1]} >>> "))
                    if t == 1 or t == 2:
                        self.order_list.append(f"{self.temp[t-1]} {self.menu[n-1]}")
                    else :
                        print("잘못된 입력입니다. 1 또는 2를 입력하세요.")
                print(f"\n고르신 메뉴는 {self.temp[t-1]} {self.menu[n-1]} : {self.price[n-1]}원 입니다.")
            elif n == 0:
                print(f"\n주문하신 메뉴는 {self.order_list}로 총 {self.price_sum}원 입니다.")
                self.pay()  # 바로 지불 메서드로 이동
            else:
                print("\n잘못된 입력입니다. 다시 입력하세요\n")

    # 지불 메서드
    def pay(self):
        p = 0  # p값 초기화
        while p != 1 and p != 2 :
            p = int(input("\n어떤 방법으로 지불하시겠습니까? 1. 현금, 2. 카드 >>> "))
            if p == 1:
                print("\n직원을 호출하겠습니다.\n주문서를 출력합니다\n")
                self.table()  # 지불 시 주문서 자동 출력
            elif p == 2:
                print("\nIC칩 방향에 맞게 카드를 꽂아주세요.\n주문서를 출력합니다\n")
                self.table()
                  # 지불 시 주문서 자동 출력
            else:
                print("\n잘못된 입력입니다. 다시 입력하세요\n")

    # 데코레이터 메서드 생성
    def deco(fn):
        def wrapper(self):
            print('⟝' + '-' * 12 + '주문서' + '-' * 12 + '⟞')
            for i in range(5):
                print('|' + ' ' * 31 + '|')
            fn(self)
            for i in range(5):
                print('|' + ' ' * 31 + '|')
            print('⟝' + '-' * 30 + '⟞')
        return wrapper

    # 주문서 출력
    @deco  # 데코레이터 적용
    def table(self):
        print(" ") # 보기 좋게 한 줄 띄우기 :)
        for i in range(len(self.order_list)):
            print(self.order_list[i], ' : ', self.order_price[i])

        print('\n합계 금액 :', self.price_sum)

a = Kiosk()  # 객체 생성
a.start()
```
