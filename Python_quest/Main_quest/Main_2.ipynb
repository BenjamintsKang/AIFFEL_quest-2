```python
import random

class Aiffel:
    # 학생 정보를 보관할 딕셔너리 생성
    students_dict = {}

    def __init__(self, name, year, generation, course):
        self.name = name
        self.year = year
        self.generation = generation
        self.course = course
        self.student_id = None  # 추후에 id 생성 후 적용 예정
        self.score = 0
        self.penalty = 0
        self.total_score = 0

    def __repr__(self):
        return f"이름 : {self.name}, 학번 : {self.student_id}, 점수 : {self.score}, 페널티 : {self.penalty}, 총점 : {self.total_score}"

    @classmethod
    def assign_student_ids(cls):
        sorted_students = sorted(cls.students_dict.values(), key=lambda student: student.name)
        for index, student in enumerate(sorted_students, start=1):
            part1 = str(student.year)[-2:]
            part2 = f"{student.generation:02d}"
            if student.course == "리서치":
                part3 = '1'
            elif student.course == "코어":
                part3 = '2'
            else:
                part3 = '0'
            part4 = f"{index:02d}"
            student.student_id = part1 + part2 + part3 + part4

    @classmethod
    def print(cls):
        for student in sorted(cls.students_dict.values(), key=lambda student: student.name):
            print(student)

    def add_score(self, score):
        self.score += score
        self.calculate_total_score()

    def add_penalty(self, penalty):
        self.penalty += penalty
        self.calculate_total_score()

    def calculate_total_score(self):
        self.total_score = self.score - self.penalty

    @classmethod
    def delete_student(self,name):
        Aiffel.students_dict.pop(name)
        return f"{name} 학생의 정보를 삭제합니다."

    @classmethod
    def grouping(cls, team_size):
        students = list(cls.students_dict.keys())
        random.shuffle(students)
        num_teams = (len(students) + team_size - 1) // team_size  # 팀 수 계산
        teams = {f"Team {i+1}": [] for i in range(num_teams)}

        # 기본 팀 구성
        for index, student in enumerate(students):
            team_number = index // team_size + 1
            if team_number > num_teams:
                team_number = team_number % num_teams
            teams[f"Team {team_number}"].append(student)

        # 남은 학생 배치
        extra_students = len(students) % team_size
        for i in range(extra_students):
            teams[f"Team {i+1}"].append(students[-(i+1)])

        for team, members in teams.items():
            print(f"{team} : {members}")


```

```python
# 점수 주기
Aiffel.students_dict['김지훈'].add_score(95)
Aiffel.students_dict['김지훈'].add_penalty(5)
# 페널티 주기
Aiffel.students_dict['이서연'].add_score(88)
Aiffel.students_dict['이서연'].add_penalty(3)
# 점수.페널티, 총점을 학번 순서대로 출력
Aiffel.print()
# 학생 정보를 삭제
Aiffel.delete_student('김지훈')
# 재출력
Aiffel.print()
# 팀 배정
Aiffel.grouping(4)
```
